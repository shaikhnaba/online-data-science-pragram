## Add column

```python
df['Attack'].mean()
```

```python
attack_mean=df['Attack'].mean()

def set_attack(val):
    if val < attack_mean:
        return'Attack low'
    else:
        return'Attack high'
```

```python
df['attck_high_low']=df['Attack'].apply(set_attack)
df
```

```python
df['HP'].mean()
```

```python
hp_mean=df['HP'].mean()

def set_hp(val):
    if val < hp_mean:
        return'HP low'
    else:
        return'HP high'
```

```python
df['hp_high_low']=df['HP'].apply(set_hp)
df
```

```python
df['Speed'].mean()
```

```python
speed_mean=df['Speed'].mean()

def set_speed(val):
    if val < speed_mean:
        return'Speed low'
    else:
        return'Speed high'
```

```python
df['speed_high_low']=df['Speed'].apply(set_speed)
df.head(20)
```
